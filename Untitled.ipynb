```python
# Importing Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Creating csv path and loading first data in Pandas DataFrame
stroke_csv = "Resources/healthcare-dataset-stroke-data.csv"
stroke_df = pd.read_csv(stroke_csv)
stroke_df
```

```python
# drop nan rows using .dropna()
stroke_df = stroke_df.dropna()
stroke_df
```

```python
# Dropping/removing unwanted columns
stroke_df.drop(["id"], axis=1, inplace=True)
stroke_df
```

```python
# do a .value_counts() to see if there is any unwanted row on gender column
stroke_df['gender'].value_counts()
```

```python
# view the other gender row
other_gender = stroke_df.loc[stroke_df["gender"] == "Other", :]
other_gender
```

```python
# Drop the other row from the dataset
stroke_df_new = stroke_df.drop(3116)

stroke_df_new
```

```python
# Do a .value_counts() to see if the unwanted row  3116 on gender column as been removed
stroke_df_new['gender'].value_counts()
```

```python
# drop the index columns, to set index in right order
stroke_df_new.reset_index(drop=True, inplace=True)
stroke_df_new
```

```python
# Pie chart showing the percentage of female to male
plotting_sex = stroke_df_new['gender'].value_counts()

colors = ['red', 'yellow']

label = ['Female', 'Male']
plt.title('Distribution of Female Versus Male')
plotting_sex.plot(kind='pie',  autopct= "%1.1f%%", figsize=(20,10), colors = colors, startangle=140, 
                   shadow = True,)

plt.savefig("./images/gender_type_pie_chart.png")
```

```python
#Pie chart showing the percentage of urban to rural residence_type
plotting_sex = stroke_df_new['Residence_type'].value_counts()

colors = ['green', 'blue']

label = ['Rural', 'Urban']
plt.title('Distribution of Urban to Rural|')
plotting_sex.plot(kind='pie',  autopct= "%1.1f%%", figsize=(20,10), colors = colors, startangle=140, 
                   shadow = True,)
plt.savefig("./images/Residence_type_pie_chart.png")
```

```python
#Pie chart showing the percentage of ever_married
plotting_sex = stroke_df_new['ever_married'].value_counts()

colors = ['grey', 'yellow']

label = ['Yes', 'No']
plt.title('Distribution of Ever Married or Never Married|')
plotting_sex.plot(kind='pie',  autopct= "%1.1f%%", figsize=(20,10), colors = colors, startangle=140, 
                   shadow = True,)
plt.savefig("./images/ever_married_pie_chart.png")
```

```python
#Pie chart showing the percentage distribution of people with stroke 
plotting_sex = stroke_df_new['stroke'].value_counts()

colors = ['grey', 'yellow']

label = ['Yes', 'No']
plt.title('Distribution of People with Stroke where 1 means stroke and 0 means no stroke')
plotting_sex.plot(kind='pie',  autopct= "%1.1f%%", figsize=(20,10), colors = colors, startangle=140, 
                   shadow = True,)
plt.savefig("./images/stroke_pie_chart.png")
```

```python
#plot a bar chart using pandas showing smoking status
smoking_plot = stroke_df_new['smoking_status'].value_counts()
smoking_plot.plot(kind='bar',figsize=(10,5), color='purple')

# set a title for the chart
plt.title ("Bar Chart of Average Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("Smoking Status count")
plt.savefig("./images/smoking_status_bar_chart.png")
```

```python
#plot a bar chart showing work type
work_type_plot = stroke_df_new['work_type'].value_counts()
work_type_plot.plot(kind='bar',figsize=(10,5), color='purple')

# set a title for the chart
plt.title ("Bar Chart Showing Work Type")
plt.xlabel("Work Type")
plt.ylabel("Work type count")
plt.savefig("./images/work_type_bar_chart.png")
```

```python
# age range
mini = stroke_df_new["age"].min()
maxi = stroke_df_new["age"].max()
print(f'The minimum age is {mini}, and the maximum age is {maxi}')
```

```python
ages = stroke_df_new["age"]
```

```python
stroke_df_newage = pd.cut(x=stroke_df_new["age"], bins=[29,39, 49,59,69,79,89], labels=['30s', '40s','50s','60s','70s','80s'])
stroke_df_newage.value_counts()

```

```python
out = pd.cut(ages, bins=[0, 35, 45, 55, 65, 75, 85], include_lowest=True)
ax = out.value_counts(sort=False).plot.bar(rot=0, color="b", figsize=(15,7))
ax.set_xticklabels([c[1:-1].replace(","," to") for c in out.cat.categories])
plt.show()
```
